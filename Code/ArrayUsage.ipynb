# 랜덤 패키지 임포트하기

```python
import random
import ctypes
```

# 사용자의 입력을 받아 들이는 부분

```python
n= int(input("몇 개의 숫자를 생성할까요?"))
```

# 리스트 형태의 배열을 생성

```python
ArrayType = ctypes.c_int * n
A = ArrayType()

for i in range(n):
  A[i] = random.randint(1, 100)
```

# 생성된 배열을 확인

```python
print(A)

A_addr = ctypes.addressof(A)
for i in range(n):
    print(A[i])
```

# 배열의 최대값 구하기

**Algorithm** *arrayMax(A, n)*<br>
&emsp;**Input** array *A* of *n* integers<br>
&emsp;**Output** maximum element of *A*<br>
<br>
&emsp;*currentMax* $\leftarrow$ *A*[0]<br>
&emsp;**for** *i* $\leftarrow$ 1 **to** *n* − 1 **do**<br>
&emsp;&emsp;**if** *A*[*i*] > *currentMax* **then**<br>
&emsp;&emsp;&emsp;*currentMax* $\leftarrow$ *A*[*i*]<br>
&emsp;**return** *currentMax*

```python
def arrayMax(A:ArrayType, n:int) -> int:
    currentMax=A[0]
    for i in range(1, n):
        if A[i] > currentMax:
            currentMax = A[i]
    return currentMax

print(arrayMax(A, n))
```

# 배열의 최대 값의 인덱스 구하기

```python
def maxArrayIndex(A:ArrayType, n:int) -> int:
    currentMax=A[0]
    maxIndex=0
    for i in range(1, n):
        if A[i] > currentMax:
            currentMax = A[i]
            maxIndex = i
    return maxIndex

print(maxArrayIndex(A, n))
```

```python
print("가장 큰 수는 ", (maxArrayIndex(A,n)+1), "번째 있습니다.")

print("가장 큰 수의 주소는 ", (A_addr + maxArrayIndex(A,n)*ctypes.sizeof(ctypes.c_int)))
print("가장 큰 수는 ", ctypes.c_int.from_address(A_addr + maxArrayIndex(A,n)*ctypes.sizeof(ctypes.c_int)).value, "입니다.")
```

# 배열 탐색

**Algorithm** *arraySearch(A,n, x)*<br>
&emsp;**Input** array *A* of *n* integers<br>
&emsp;**Output** location of *x* in *A*<br>
<br>
&emsp;**for** *i* $\leftarrow$ 0 **to** *n* − 1 **do**<br>
&emsp;&emsp;**if** *A*[*i*] == *x* **then**<br>
&emsp;&emsp;&emsp;**return** *i*<br>
&emsp;**return** -1

```python
def arraySearch(A:ArrayType, n:int, x:int) -> int:
  for i in range(n):
    if A[i]==x:
      return i
  return -1
```

```python
f=arraySearch(A,n,2)
if f>0:
  print("찾는 숫자는 ", (f+1), "번째 있습니다.")
  print("찾는 숫자의 주소는 ", (A_addr + f*ctypes.sizeof(ctypes.c_int)), "입니다.")
else:
  print("찻는 숫자는 배열 안에 없습니다.")
```
