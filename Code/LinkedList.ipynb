## DNode 구조

```
 ┌──────────┬──────────┬──────────┐
 │   prev   │   data   │   next   │
 │  (이전)   │  (원소)   │  (다음)   │
 └──────────┴──────────┴──────────┘
```

## 연결 구조

```
          ┌──────┐    ┌──────┐    ┌──────┐    ┌──────┐    ┌─────────┐
          │ head │───>│  A   │───>│  B   │───>│  C   │───>│ trailer │
          │      │<───│      │<───│      │<───│      │<───│         │
          └──────┘    └──────┘    └──────┘    └──────┘    └─────────┘
           센티널        노드1       노드2       노드3       센티널
```

- **head**: 시작 센티널 노드 (데이터 저장 X)
- **trailer**: 끝 센티널 노드 (데이터 저장 X)
- 각 노드는 **prev**(이전 노드)와 **next**(다음 노드)를 가리키는 양방향 연결

````python
class DNode:
  def __init__(self, d, p, n): #생성자, d: element, p: previous, n:next
    self.data=d
    self.prev=p
    self.next=n

  def getPrev(self):
    return self.prev

  def setPrev(self, p):
    self.prev=p

  def getNext(self):
    return self.next

  def setNext(self, n):
    self.next=n

  def getElement(self):
    return self.data

  def setElement(self, d):
    self.data=d
````

````python
class LinkedList:

  def __init__(self):
    self.head=DNode("head",None, None)
    self.trailer=DNode("trailer",self.head,None)
    self.head.setNext(self.trailer)
    self.numElem=0

  def isEmpty(self):
        if(self.numElem==0):
            return True
        else:
            return False

  def size(self):
        #return self.numElem
        size=0
        cur=self.head
        while not cur.getNext() == self.trailer:
            cur=cur.getNext()
            size=size+1
        return size

  def First(self):
    if self.numElem==0:
      print("원소가 없습니다.")
    else:
      return self.head.getNext()

  def Last(self):
    if self.numElem==0:
      print("원소가 없습니다.")
    else:
      return self.trailer.getPrev()


  def insertFirst(self,elem):
    v=DNode(elem, self.head, self.head.getNext())
    (self.head.getNext()).setPrev(v)
    self.head.setNext(v)
    self.numElem=self.numElem+1
    return v

  def insertLast(self,elem):
    v=DNode(elem, self.trailer.getPrev(), self.trailer)
    (self.trailer.getPrev()).setNext(v)
    self.trailer.setPrev(v)
    self.numElem=self.numElem+1
    return v

  def insertAfter(self, pos, elem):
    v=DNode(elem, pos, pos.getNext())
    (pos.getNext()).setPrev(v)
    pos.setNext(v)
    self.numElem=self.numElem+1
    return v

  def insertBefore(self, pos, elem):
    v=DNode(elem, pos.getPrev(), pos)
    (pos.getPrev()).setNext(v)
    pos.setPrev(v)
    self.numElem=self.numElem+1
    return v

  def remove(self, pos):
    elem=pos.getElement()
    pos.getPrev().setNext(pos.getNext())
    pos.getNext().setPrev(pos.getPrev())
    pos.setPrev(None)
    pos.setNext(None)
    self.numElem=self.numElem-1
    return elem

  def replaceElement(self, pos, elem):
    e=pos.getElement()
    pos.setElement(elem)
    return e

  def __str__(self):
    result='['
    cur=self.head.getNext()
    while cur!=self.trailer:
      if cur.getNext()!=self.trailer:
        result+=str(cur.data)+'->'
      else:
        result+=str(cur.data)
      cur=cur.getNext()
    result+=']'
    return result
````

````python
ll = LinkedList()

# ── 1. isEmpty() : 빈 리스트 확인 ──
print("=== isEmpty ===")
print(f"비어있나? {ll.isEmpty()}")  # True
print(f"리스트: {ll}")
print()

# ── 2. insertFirst() : 맨 앞에 삽입 ──
print("=== insertFirst ===")
p1 = ll.insertFirst("A")
print(f"A 삽입 → {ll}")
p2 = ll.insertFirst("B")
print(f"B 삽입 → {ll}")
p3 = ll.insertFirst("C")
print(f"C 삽입 → {ll}")
print()

# ── 3. insertLast() : 맨 뒤에 삽입 ──
print("=== insertLast ===")
p4 = ll.insertLast("D")
print(f"D 삽입 → {ll}")
p5 = ll.insertLast("E")
print(f"E 삽입 → {ll}")
print()

# ── 4. insertAfter() : 특정 노드 뒤에 삽입 ──
print("=== insertAfter ===")
print(f"현재: {ll}")
p6 = ll.insertAfter(p1, "A-1")   # A 뒤에 A-1 삽입
print(f"A 뒤에 A-1 삽입 → {ll}")
print()

# ── 5. insertBefore() : 특정 노드 앞에 삽입 ──
print("=== insertBefore ===")
p7 = ll.insertBefore(p4, "C-D")  # D 앞에 C-D 삽입
print(f"D 앞에 C-D 삽입 → {ll}")
print()

# ── 6. size() : 리스트 크기 확인 ──
print("=== size ===")
print(f"크기: {ll.size()}")
print(f"비어있나? {ll.isEmpty()}")  # False
print()

# ── 7. First() / Last() : 첫번째, 마지막 원소 ──
print("=== First / Last ===")
print(f"첫 번째 원소: {ll.First().getElement()}")
print(f"마지막 원소: {ll.Last().getElement()}")
print()

# ── 8. replaceElement() : 원소 교체 ──
print("=== replaceElement ===")
print(f"교체 전: {ll}")
old = ll.replaceElement(p1, "A★")  # A → A★
print(f"A를 A★로 교체 (이전값: {old}) → {ll}")
print()

# ── 9. remove() : 노드 삭제 ──
print("=== remove ===")
print(f"삭제 전: {ll}")
removed = ll.remove(p6)            # A-1 삭제
print(f"A-1 삭제 (삭제값: {removed}) → {ll}")
removed = ll.remove(p3)            # C 삭제 (맨 앞)
print(f"C 삭제 (삭제값: {removed}) → {ll}")
removed = ll.remove(p5)            # E 삭제 (맨 뒤)
print(f"E 삭제 (삭제값: {removed}) → {ll}")
print()

# ── 10. 최종 상태 ──
print("=== 최종 상태 ===")
print(f"리스트: {ll}")
print(f"크기: {ll.size()}")
print(f"첫 번째: {ll.First().getElement()}")
print(f"마지막: {ll.Last().getElement()}")
````
