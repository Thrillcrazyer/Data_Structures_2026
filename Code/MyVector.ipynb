```python
import ctypes

class MyVector:
    def __init__(self, d):
        if isinstance(d, int):
            self.capacity = d
            self.coords = (ctypes.c_int * d)()
            self.size = 0
        else:
            raise TypeError("정수형으로 Vector의 입력값을 넣어주세요")

    def _resize(self):
        new_capacity = self.capacity * 2
        new_coords = (ctypes.c_int * new_capacity)()
        for j in range(self.size):
            new_coords[j] = self.coords[j]
        self.coords = new_coords
        self.capacity = new_capacity

    def __str__(self):
        result = '['
        for i in range(self.size):
            if i == self.size - 1:
                result += str(self.coords[i]) + ']'
            else:
                result += str(self.coords[i]) + ','
        return result

    def elementAt(self, i):
        if self.size != 0 and 0 <= i < self.size:
            return self.coords[i]
        else:
            print("요청한 범위가 올바르지 않습니다.")

    def insertAt(self, i, val):
        if self.size == self.capacity - 1:
            print("꽉 찼어요", self.size)
            self._resize()
        if 0 <= i <= self.size:
            for k in range(self.size, i, -1):
                self.coords[k] = self.coords[k - 1]
            self.coords[i] = val
            self.size += 1

    def insertLast(self, val):
        if self.size == self.capacity - 1:
            print("꽉 찼어요.", val)
            self._resize()
        self.coords[self.size] = val
        self.size += 1

    def removeAt(self, i):
        if self.size == 0:
            print("빈 벡터입니다.")
        else:
            if i < 0 or i >= self.size:
                print("잘못된 범위를 요청하였습니다.")
            else:
                ret = self.coords[i]
                for k in range(i, self.size - 1):
                    self.coords[k] = self.coords[k + 1]
                self.size -= 1
                return ret

    def replaceAt(self, i, val):
        if i < 0 or i >= self.size:
            print("잘못된 범위를 요청하였습니다.")
        else:
            ret = self.coords[i]
            self.coords[i] = val
            return ret

if __name__ == '__main__':
    mv = MyVector(2)

    for i in range(100):
        mv.insertAt(i, i)
    print(mv.removeAt(50))
    print(mv.replaceAt(0, 50))

    print(mv)
```

```python
class MyVector:
  def __init__(self, d): #initialize(초기화, 입력받는 것)
    if isinstance(d, int):
      self.coords=[0]*d #실제 데이터가 저장되는 공간
      self.size=0 #벡터의 사이즈.
    else:
      try: #실행하는 문장
        self.coords=[val for val in d]
      except TypeError: #제외시키는 문장
        raise TypeError('invalid parameter type')

  def elementAt(self, i):
    if self.size!=0 and i>=0 and i<self.size:
      return self.coords[i]
    else:
      print("요청한 범위가 올바르지 않습니다.")

  def insertAt(self, i, val): #인덱스 i 위치에 val값을 넣어라.
    if self.size==len(self.coords)-1:
      print("꽉 찼어요", self.size)
      newList=[0]*(len(self.coords)*2)
      for j in range(0, self.size):
        newList[j]=self.coords[j]
      self.coords=newList
    if i>=0 and i<=self.size:
      for k in range(self.size, i, -1):
        self.coords[k]=self.coords[k-1] #k값을 이전 k-1값으로 채운다.
      self.coords[i]=val
      self.size=self.size+1

  def insertLast(self, val):
    if self.size==len(self.coords)-1:
      print("꽉 찼어요.",val)
      newList=[0]*(len(self.coords)*2)
      for j in range(0, self.size):
        newList[j]=self.coords[j]
      self.coords=newList
    self.coords[self.size]=val
    self.size=self.size+1

  def removeAt(self, i):
    if self.size==0:
      print("빈 벡터입니다.")
    else:
      if i<0 and i>=self.size:
        print("잘못된 범위를 요청하였습니다.")
      else:
        ret=self.coords[i] #지우기 이전 값을 저장
        for k in range(i, self.size):
          self.coords[k]=self.coords[k+1]
        self.size=self.size-1
        return ret #지우기 이전 값을 리턴

  def replaceAt(self, i, val):
    if i<0 or i>=self.size:
      print("잘못된 범위를 요청하였습니다.")
    else:
      ret=self.coords[i] #지우기 이전 값을 저장
      self.coords[i]=val
      return ret #지우기 이전 값을 리턴

  def printVector(self):
    print('[',end='')
    for i in range(0, self.size):
      if i==self.size-1:
        print(self.coords[i],']')
      else:
        print(self.coords[i], ',', end='')

if __name__=='__main__': #인터프리터에 의해 실행이 되는 경우, 해당 구문이 사용된 파이썬 파일을 직접 실행했을 때만 실행시킨다.
  mv=MyVector(2)

  #mv.insertAt(0,0)
  #mv.insertAt(1,1)
  #mv.insertAt(2,2)

  #print(mv.coords)
  for i in range(0,100):
    mv.insertAt(i,i)
  print(mv.removeAt(50))
  print(mv.replaceAt(0,50)) #인덱스 0에 '50'이라는 값을 넣음

  mv.printVector()


```
