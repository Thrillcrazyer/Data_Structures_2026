```python
import ctypes

class MyVector:
    def __init__(self, d):
        if isinstance(d, int):
            self.capacity = d
            self.coords = (ctypes.c_int * d)()
            self.size = 0
        else:
            raise TypeError("정수형으로 Vector의 입력값을 넣어주세요")
    
    def __str__(self):
        result = '['
        for i in range(self.size):
            if i == self.size - 1:
                result += str(self.coords[i]) + ']'
            else:
                result += str(self.coords[i]) + ','
        return result

    def _resize(self):
        new_capacity = self.capacity * 2
        new_coords = (ctypes.c_int * new_capacity)()
        for j in range(self.size):
            new_coords[j] = self.coords[j]
        self.coords = new_coords
        self.capacity = new_capacity

    def elementAt(self, i):
        if self.size != 0 and 0 <= i < self.size:
            return self.coords[i]
        else:
            print("요청한 범위가 올바르지 않습니다.")

    def insertAt(self, i, val):
        if self.size == self.capacity - 1:
            print("꽉 찼어요", self.size)
            self._resize()
        if 0 <= i <= self.size:
            for k in range(self.size, i, -1):
                self.coords[k] = self.coords[k - 1]
            self.coords[i] = val
            self.size += 1

    def insertLast(self, val):
        if self.size == self.capacity - 1:
            print("꽉 찼어요.", val)
            self._resize()
        self.coords[self.size] = val
        self.size += 1

    def removeAt(self, i):
        if self.size == 0:
            print("빈 벡터입니다.")
        else:
            if i < 0 or i >= self.size:
                print("잘못된 범위를 요청하였습니다.")
            else:
                ret = self.coords[i]
                for k in range(i, self.size - 1):
                    self.coords[k] = self.coords[k + 1]
                self.size -= 1
                return ret

    def replaceAt(self, i, val):
        if i < 0 or i >= self.size:
            print("잘못된 범위를 요청하였습니다.")
        else:
            ret = self.coords[i]
            self.coords[i] = val
            return ret

if __name__ == '__main__':
    mv = MyVector(2)

    for i in range(100):
        mv.insertAt(i, i)
    print(mv.removeAt(50))
    print(mv.replaceAt(0, 50))

    print(mv)
```
